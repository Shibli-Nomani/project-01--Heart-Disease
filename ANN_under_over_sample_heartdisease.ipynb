```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
path='/content/gdrive/MyDrive/Final Project/train.csv'
df=pd.read_csv(path)
df.head(5)
```

```python
from sklearn.model_selection import train_test_split

import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error,confusion_matrix, precision_score, recall_score, f1_score, auc,roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import mean_absolute_error

import keras
from keras.models import Sequential
from keras.layers import Dense

```

```python
#target
y=df['HeartDisease']
#features
X=df.copy()
X=df.drop('HeartDisease', axis=1)
X.head(5)
```

#**Under Sampling**

```python
df1=df.copy()
#Target
y_us=df1['HeartDisease']
#Features

X_us=df1.copy()
X_us=df1.drop('HeartDisease', axis=1)
```

```python
from imblearn.under_sampling import NearMiss
NearMiss_obj = NearMiss()
X_us , y_us = NearMiss_obj.fit_resample(X_us,y_us)
print(X_us.shape)
print(y_us.shape)
```

```python
X_us_train, X_us_test, y_us_train, y_us_test = train_test_split(X_us,y_us,test_size = 0.2, random_state=42)
```

**Multiple Classification Model List**

```python
#to add 1st hidden layer
ml10 = Sequential()
ml10.add(Dense(6,activation = 'relu', input_dim = 17)) #no of feature 17
#to add 2nd hidden layer
ml10.add(Dense(6, activation = 'relu'))
#add output layer as target is in binary formate and reqire 1 neuron
ml10.add(Dense(1, activation = 'sigmoid'))
#compile
ml10.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ml10.fit(X_us_train, y_us_train, batch_size = 10, epochs = 50)
```

```python
 y_us_pred=ml10.predict(X_us_test)
 y_us_pred[y_us_pred <= 0.5] = 0
 y_us_pred[y_us_pred > 0.5] = 1
```

```python
fp, tp, tn = roc_curve(y_us_test, y_us_pred)

Train_Accuracy_loss = ml10.evaluate(X_us_train, y_us_train)
Train_Accuracy = (Train_Accuracy_loss[1])
Train_loss = (Train_Accuracy_loss[0])

Test_Accuracy_loss =  ml10.evaluate(X_us_test, y_us_test)
Test_Accuracy = (Test_Accuracy_loss[1])
Test_loss = (Test_Accuracy_loss[0])

Precission = precision_score(y_us_test, y_us_pred)
Recall= recall_score(y_us_test, y_us_pred)
f1_Score=f1_score(y_us_test, y_us_pred)
AUC = auc(fp, tp)
mbe=mean_absolute_error(y_us_test, y_us_pred)
list_name = [["Train Accuracy", Train_Accuracy], ["Train loss", Train_loss], ["Test Accuracy", Test_Accuracy], ["Test loss", Test_loss],
             ["Precision",Precission], ["Recall", Recall],
             ['F1 Score',f1_Score], ["AUC", AUC],["mean abs error",mbe]]
model_columns = pd.DataFrame (list_name, columns = ['measures', 'values'])
model_columns
```

```python
model_columns.transpose().reset_index().rename(columns={'index':'ANN'})
```

#**Over Sampling with SMOTETomek**

```python
df.head(5)
```

```python
df1=df.copy()
#Target
y=df1['HeartDisease']
#Features

X=df1.copy()
X=df1.drop('HeartDisease', axis=1)
```

```python
!pip install imblearn
from imblearn.combine import SMOTETomek
sm = SMOTETomek(random_state=42)
```

```python
X_sm, y_sm = sm.fit_resample(X,y)
```

```python
from sklearn.model_selection import train_test_split
X_sm_train, X_sm_test, y_sm_train, y_sm_test=train_test_split(X_sm,y_sm, test_size=0.3, random_state=42)
print(X_sm.shape)
print(y_sm.shape)
```

**ANN Model Prepare**

```python
#to add 1st hidden layer
ml10 = Sequential()
ml10.add(Dense(6,activation = 'relu', input_dim = 17)) #no of feature 17
#to add 2nd hidden layer
ml10.add(Dense(6, activation = 'relu'))
#add output layer as target is in binary formate and reqire 1 neuron
ml10.add(Dense(1, activation = 'sigmoid'))
#compile
ml10.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ml10.fit(X_sm_train, y_sm_train, batch_size = 10, epochs = 50)
```

```python
 y_sm_pred=ml10.predict(X_sm_test)
 y_sm_pred[y_sm_pred <= 0.5] = 0
 y_sm_pred[y_sm_pred > 0.5] = 1
```

```python
fp, tp, tn = roc_curve(y_sm_test, y_sm_pred)

Train_Accuracy_loss = ml10.evaluate(X_sm_train, y_sm_train)
Train_Accuracy = (Train_Accuracy_loss[1])
Train_loss = (Train_Accuracy_loss[0])

Test_Accuracy_loss =  ml10.evaluate(X_sm_test, y_sm_test)
Test_Accuracy = (Test_Accuracy_loss[1])
Test_loss = (Test_Accuracy_loss[0])

Precission = precision_score(y_sm_test, y_sm_pred)
Recall= recall_score(y_sm_test, y_sm_pred)
f1_Score=f1_score(y_sm_test, y_sm_pred)
AUC = auc(fp, tp)
mbe=mean_absolute_error(y_sm_test, y_sm_pred)
list_name = [["Train Accuracy", Train_Accuracy], ["Train loss", Train_loss], ["Test Accuracy", Test_Accuracy], ["Test loss", Test_loss],
             ["Precision",Precission], ["Recall", Recall],
             ['F1 Score',f1_Score], ["AUC", AUC],["mean abs error",mbe]]
results = pd.DataFrame (list_name, columns = ['measures', 'values'])
results
```
